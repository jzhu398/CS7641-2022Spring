```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
######################### Data Preparation #####################
df = pd.read_csv('pi_small.csv',sep = ',')
df1 = pd.read_csv('VI_small.csv',sep = ',')
df2 = pd.read_csv('Q_small.csv',sep = ',')
```

```python
###########################  small ######################
plt.figure()
plt.title('PI Reward')
plt.xlabel('Iteration')
plt.ylabel('Reward')
plt.grid(linestyle='dotted')
plt.plot(df['Reward'], color = 'red', label = 'Policy Iteration')
plt.legend()
plt.margins(0,0.1)
plt.savefig('pi_small.png')
plt.show()
plt.close()

plt.figure()
plt.title('VI Reward')
plt.xlabel('Iteration')
plt.ylabel('Reward')
plt.grid(linestyle='dotted')
plt.plot(df1['Reward'],color = 'yellow', label = 'Value Iteration')
plt.legend()
plt.margins(0,0.1)
plt.savefig('vi_small.png')
plt.show()
plt.close()


plt.figure()
plt.title('Q-Learning Reward')
plt.xlabel('Iteration')
plt.ylabel('Reward')
plt.grid(linestyle='dotted')
plt.plot(df2['Reward'],color = 'green', label = 'Q-Learning')
plt.legend()
plt.margins(0,0.1)
plt.savefig('Q_small.png')
plt.show()
plt.close()
```
